```r
suppressMessages({library(ggplot2)
library(gganimate)
library(babynames)
library(hrbrthemes)})
```

```r
indices <- as.data.frame(read.csv('C:/Users/User/Desktop/indices_timeseries.csv', header=T)) #Load excel file
colnames(indices)[1] <- 'doy'
```

```r
ndvi <- indices[1:2]
ndvi <- cbind(ndvi,rep('NDVI',nrow(ndvi)))
colnames(ndvi) <- c('Day_of_Year','n','index')

ndwi <- indices[c(1,3)]
ndwi <- cbind(ndwi,rep('NDWI',nrow(ndwi)))
colnames(ndwi) <- c('Day_of_Year','n','index')

psri <- indices[c(1,4)]
psri <- cbind(psri,rep('PSRI',nrow(psri)))
colnames(psri) <- c('Day_of_Year','n','index')
```

```r
indices <- rbind(ndvi,ndwi,psri)
indices <- indices[order(indices$Day_of_Year),]
rownames(indices) <- c(1:nrow(indices))
indices <- indices[c(1,3,2)]
```

```r
# Plot
suppressMessages({indices %>%
  ggplot( aes(x=Day_of_Year, y=n, group=index, color=index)) +
    geom_line() +              
    geom_point(aes(group = seq_along(Day_of_Year))) +
    scale_colour_manual(values = c("palegreen3", "lightskyblue3", "khaki2")) +
    ggtitle("NDVI, NDWI and PSRI Time-series") +
    theme_ipsum() +
    ylab("Index Value") +
    xlab("Day of Year (DOY)") +
    transition_reveal(Day_of_Year)})
```
